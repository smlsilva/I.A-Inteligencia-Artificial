**HISTOGRAMA IMAGEM 6 E IMAGEM 1**

```
img1 = cv.imread('6fb5d5fc1d588a0f1a75722e080af28d.png')
img1 = cv.cvtColor(img1, cv.COLOR_BGR2RGB)
img6 = cv.bitwise_not(img1)

hist = cv.calcHist(img1,[0], None, [256], [0,256])
hist1 = cv.calcHist(img6, [0], None, [256], [0,256])

plt.figure(figsize=(40,20))
plt.subplot(241), plt.plot(hist,'--', color= 'red'), plt.xlim(0, 256)
plt.xlabel('Imagem_1')

plt.subplot(242), plt.plot(hist1,'--', color= 'blue'), plt.xlim(0, 256)
plt.xlabel('Imagem_6')


plt.show()
```
