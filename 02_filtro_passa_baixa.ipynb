**IMPORTANDO AS LIBRARIES E FAZENDO O PROCESSAMENTO DAS MESMAS**

```
import cv2 as cv
import matplotlib.pyplot as plt
import random as rd

img = cv.imread("modelo.jpg")
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

prob = 0.05 # DETERMINEI A PROBABILIDADE | QUANTO MAIOR, MAIS ESCURO FICA A IMAGEM
res = 1 - prob # PARA FAZER A VARIAÇÃO DOS PIXELS
for i in range(img.shape[0]):
  for j in range(img.shape[1]):
    if rd.random() < prob:
       img[i][j] = 0
    elif rd.random() > res:
       img[i][j] = 255
    else:
       img[i][j] = img[i][j]

img1 = cv.medianBlur(img, 11) # FILTRAGEM MÉDIA
img2 = cv.GaussianBlur(img, (13,13), 0) # FILTRAGEM GAUSSIANA

plt.figure(figsize=(20,15))
plt.subplot(241), plt.imshow(img), plt.title('Sal e Pimenta')
plt.xticks([])
plt.yticks([])

plt.subplot(242), plt.imshow(img1), plt.title('MedianBlur | Média')
plt.xticks([])
plt.yticks([])

plt.subplot(243), plt.imshow(img2), plt.title('GaussianBlur | Gaussiana')
plt.xticks([])
plt.yticks([])

plt.show()
```
