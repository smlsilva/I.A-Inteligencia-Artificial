```
import cv2 as cv
import matplotlib.pyplot as plt
```

```
!wget -O 'meninas.jpg' 'https://jornalismojunior.com.br/jpress_old/wp-content/uploads/2014/08/dg.jpg'
```

```
img = cv.imread('meninas.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
img = cv.cvtColor(img, cv.COLOR_RGB2GRAY)

img1 = cv.Sobel(img,cv.CV_8U,1,0,ksize=5)
img2 = cv.Sobel(img,cv.CV_64F,0,1,ksize=5)
img3 = cv.Laplacian(img,cv.CV_32S,ksize=5)
img4 = cv.GaussianBlur(img,(5,5),cv.BORDER_ISOLATED)

plt.figure(figsize=(20,15))
plt.subplot(241), plt.imshow(img1, cmap='gray')
plt.xticks([])
plt.yticks([])

plt.subplot(242), plt.imshow(img2, cmap='gray')
plt.xticks([])
plt.yticks([])

plt.subplot(243), plt.imshow(img3, cmap='gray')
plt.xticks([])
plt.yticks([])


plt.subplot(244), plt.imshow(img4, cmap='gray')
plt.xticks([])
plt.yticks([])

plt.show()
```
